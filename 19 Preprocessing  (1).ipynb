```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder

df = pd.read_csv('AIR_QUALITY_DATASET.xlsx.csv')

print("Original Shape:", df.shape)

df = df.drop(columns=['StationId', 'Datetime'])

num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(include='object').columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

if 'AQI_Bucket' in cat_cols:
    le = LabelEncoder()
    df['AQI_Bucket'] = le.fit_transform(df['AQI_Bucket'])

df = df.drop_duplicates()

Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[num_cols] < (Q1 - 1.5 * IQR)) |
          (df[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

scaler = StandardScaler()

X = df.drop('AQI_Bucket', axis=1)
y = df['AQI_Bucket']

X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print("Final Shape:", X_scaled.shape)

X_scaled.to_csv('X_final.csv', index=False)
y.to_csv('y_final.csv', index=False)
```
